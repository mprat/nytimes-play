```python
import requests
base_uri = "http://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
apikeyfile=open('api-key.txt', 'rb')
allkeys = list(apikeyfile)
apikeyfile.close()
api_key = [keypair.split("++")[1] for keypair in allkeys if "articlesearch" in keypair][0]
```

```python
begin_date = "20110130"
end_date = "20120130"
query = "middle east"
fq = 'source:("The New York Times")'
vals = {'begin_date':begin_date, 'end_date': end_date,'q':query, 'fq': fq, 'api-key': api_key}
```

```python
r = requests.get(base_uri, params=vals)
```

```python
rjson = r.json()
print "numhits = ", rjson['response']['meta']['hits']
print rjson['response']['docs']
```

```python
import requests
from bs4 import BeautifulSoup

base_url = 'http://www.nytimes.com'
r = requests.get(base_url)
soup = BeautifulSoup(r.text)
print soup
```

```python
for story_heading in soup.find_all(class_="story-heading"): 
    if story_heading.a: 
        print story_heading.a.text.replace("\n", " ").strip()
    else: 
        print story_heading.contents[0].strip()
    
```
